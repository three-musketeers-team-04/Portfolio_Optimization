```python
start_sp = datetime.datetime(2009, 10, 1)
end_sp = datetime.datetime(2019, 10, 1)
```

```python
stock_list = ['GOOG','FB', 'AAPL', 'AMZ', 'MSFT']
close_price = {}

for stock in stock_list:
    close=[]
    result = pdr.get_data_yahoo(stock,start_sp, end_sp)['Close']
    close.append(result)
    close_price[stock] = close

print(close_price)
```
